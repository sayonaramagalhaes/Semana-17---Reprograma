```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv('/content/Relatório de Transplantes Realizados (Rio Grande do Norte) - Evolução 2001 - 2023 - Table 1.csv')
```

```python
df.head(22)
```

```python
#renomear colunas
df.rename(columns={'Unnamed: 0':'Ano'},inplace=True)
df.rename(columns={'Unnamed: 1':'Coração'},inplace=True)
#remover colunas
df.drop(columns=['Unnamed: 2'],inplace=True)
#remomer linha
#df.drop(index=0,inplace=True)
```

```python
df.rename(columns={'Unnamed: 3':'Fígado'},inplace=True)
df.rename(columns={'Unnamed: 4':'Figado vivo'},inplace=True)
df.rename(columns={'Unnamed: 5':'Figado falecido'},inplace=True)
df.rename(columns={'Unnamed: 6':'Pancreas'},inplace=True)
df.rename(columns={'Unnamed: 7':'Pulmão'},inplace=True)
```

```python
from os import rename
df.rename(columns={'Unnamed: 8':'Pulmão vivo'},inplace=True)
df.rename(columns={'Unnamed: 9':'Pulmão falecido'},inplace=True)
df.rename(columns={'Unnamed: 10':'Rim'},inplace=True)
df.rename(columns={'Unnamed: 11':'Rim vivo'},inplace=True)
df.rename(columns={'Unnamed: 12':'Rim falecido'},inplace=True)
df.rename(columns={'Unnamed: 13':'Pancreas rim'},inplace=True)
df.rename(columns={'Unnamed: 14':'Intestino Isolado'},inplace =True)
df.rename(columns={'Unnamed: 15':'Multivisceral'},inplace =True)
df.rename(columns={'Unnamed: 16':'Total Orgãos'},inplace =True)
df.rename(columns={'Unnamed: 17':'Córnea'},inplace =True)
df.rename(columns={'Unnamed: 18':'Médula óssea'},inplace =True)
df.rename(columns={'Unnamed: 19':'MO Autólogo'},inplace =True)
df.rename(columns={'Unnamed: 20':'Mo Aparentado'},inplace =True)
df.rename(columns={'Unnamed: 21':'Mo Nap'},inplace =True)
df.rename(columns={'Unnamed: 22':'Total Geral'},inplace =True)



```

```python
#remover linha
df.drop(index=0,inplace=True)
```

```python
#grafico ano rim
sns.lineplot(x='Ano',y='Rim',data=df)
plt.show()
```

```python
#grafico total orgaos ano
sns.lineplot(x='Ano',y='Total Orgãos',data=df)
plt.show()
```

```python
#grafico ano com mais orgaos
sns.barplot(x='Ano',y='Total Orgãos',data=df)
plt.show()
```

```python
#grafico todos os orgaos
sns.lineplot(x='Ano',y='Total Geral',data=df)
plt.show()
```

```python
#salvar dataset
df.to_csv('Relatório de Transplantes Realizados (Rio Grande do Norte) - Evolução 2001 - 2023 - Table 1.csv',index=False)
```
