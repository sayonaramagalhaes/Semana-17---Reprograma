```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Relatório de Doação (Rio Grande do Norte) - Evolução 2001 - 2023 - Table 1 (2).csv')
df.head()
```

```python
#remover linha 0
df=df.drop(0)
```

```python
df.head(15)
```

```python
#renomer linhas
df=df.rename(columns={'Unnamed: 0':'Ano'})
df=df.rename(columns={'Unnamed: 1':'Potencial Doador'})
df=df.rename(columns={'Unnamed: 2':'Potencial Doador (PMP)'})
df=df.rename(columns={'Unnamed: 3':'Doador Efetivo'})
df=df.rename(columns={'Unnamed: 4':'Doador Efetivo (PMP)'})
df=df.rename(columns={'Unnamed: 5':'Percentual de Efetivação'})
df=df.rename(columns={'Unnamed: 6':'Entrevista Familiar'})
df=df.rename(columns={'Unnamed: 7':'Negativa Familiar'})
df=df.rename(columns={'Unnamed: 8':'Negativa Familiar (%)'})
```

```python
#remover linha 1
df=df.drop(1)
```

```python
df.describe()
```

```python
#salvar dataset
df.to_csv('Relatório de Doação (Rio Grande do Norte) - Evolução 2001 - 2023 - Table 1 (2).csv', index=False)
```
